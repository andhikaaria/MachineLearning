```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Dataset/Amazon Business Research Analyst Dataset.csv')
```

```python
data.info()
```

```python
# Menghitung jumlah nilai null dalam setiap kolom
jumlah_null_per_kolom = data.isnull().sum()

# Menampilkan hasil
print("Jumlah nilai null dalam setiap kolom:")
print(jumlah_null_per_kolom)
```

```python
# Menghapus karakter underscore (_) dari nama kolom
data.columns = data.columns.str.replace('_', ' ')

# Menampilkan DataFrame setelah perubahan
data
```

```python
# pip install feature-engine
```

```python
from feature_engine.imputation import RandomSampleImputer

# Inisialisasi RandomSampleImputer
random_imputer = RandomSampleImputer(variables=['Delivery person Age'], random_state=42)

# Fit dan transform data
data = random_imputer.fit_transform(data)

# Menghitung mean dari kolom "Delivery person Ratings"
mean_delivery_ratings = data['Delivery person Ratings'].mean()

# Mengisi nilai kosong dengan mean
data['Delivery person Ratings'].fillna(mean_delivery_ratings, inplace=True)

# Menampilkan DataFrame setelah transformasi
data
```

```python
data.shape
```

```python
# Menampilkan nilai unik pada kolom "Type of vehicle"
unique_Type_of_vehicle = data['Type of vehicle'].unique()

# Menampilkan nilai unik
print("Nilai Unik pada Kolom 'Type of vehicle':")
print(unique_Type_of_vehicle)
```

```python
# Menampilkan nilai unik pada kolom "Road traffic density"
unique_traffic_density = data['Road traffic density'].unique()

# Menampilkan nilai unik
print("Nilai Unik pada Kolom 'Road traffic density':")
print(unique_traffic_density)
```

```python
# Menghitung modus dari kolom "Road traffic density"
modus_traffic_density = data['Road traffic density'].mode()[0]

# Mengganti nilai NaN dengan modus
data['Road traffic density'].fillna(modus_traffic_density, inplace=True)

# Menampilkan DataFrame setelah perubahan
data
```

```python
# Membuat peta nilai yang akan digunakan untuk penggantian
type_mapping = {
    'electric_scooter': 1,
    'motorcycle': 2,
    'scooter': 3,
    'bicycle': 4
}

# Menggantikan nilai dalam kolom "Type of vehicle" dengan menggunakan peta
data['Type encoded vehicle'] = data['Type of vehicle'].map(type_mapping)

# Menampilkan DataFrame setelah penambahan kolom baru
data
```

```python
# Membuat peta nilai yang akan digunakan untuk penggantian
traffic_density_mapping = {
    'Low': 1,
    'Jam': 2,
    'Medium': 3,
    'High': 4
}

# Menggantikan nilai dalam kolom "Road traffic density" dengan menggunakan peta
data['Traffic density encoded'] = data['Road traffic density'].map(traffic_density_mapping)
```

```python
# Select subset of predictors
cols_to_use = ['Traffic density encoded', 'Delivery person Ratings', 'Type encoded vehicle']
X = data[cols_to_use]
```

```python
from sklearn.model_selection import train_test_split
# Select target
y = data['Delivery person Age']

# Separate data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42, test_size=0.3)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
my_model_0 = XGBRegressor(n_estimators=500)
my_model_0.fit(X_train, y_train)
```

```python
my_model_1 = XGBRegressor(n_estimators=110)
my_model_1.fit(X_train, y_train)
```

```python
predictions_0= my_model_0.predict(X_valid)
print("Mean Absoulute Error : " + str(mean_absolute_error(predictions_0, y_valid)))
```

```python
predictions_1 = my_model_1.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(predictions_1, y_valid)))
```

```python
my_model_1 = XGBRegressor(n_estimators=110)
my_model_1.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Tentukan model XGBoost
model_grid = XGBRegressor()

# Tentukan parameter grid
param_grid = {
    'n_estimators': [150, 200, 250, 300, 1000]  # Ganti dengan rentang nilai yang ingin Anda coba
}

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(estimator=model_grid, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error')

# Lakukan fitting menggunakan data training
grid_search.fit(X_train, y_train)

# Dapatkan parameter terbaik
best_n_estimators = grid_search.best_params_['n_estimators']

print("Nilai n_estimators terbaik:", best_n_estimators)
```

```python
# Tentukan range nilai n_estimators yang akan diuji
n_estimators_list = [100, 150, 200, 250, 300]

# Inisialisasi sebuah list untuk menyimpan hasil MAE
mae_results = []

# Loop melalui setiap nilai n_estimators
for n_estimators in n_estimators_list:
    # Inisialisasi model dengan n_estimators tertentu
    model_grid = XGBRegressor(n_estimators=n_estimators)

    # Fitting model menggunakan data training
    model_grid.fit(X_train, y_train)

    # Lakukan prediksi pada data validasi
    predictions = model_grid.predict(X_valid)

    # Hitung Mean Absolute Error
    mae = mean_absolute_error(predictions, y_valid)

    # Tampilkan nilai n_estimators dan MAE
    print(f"n_estimators: {n_estimators}, Mean Absolute Error: {mae}")

    # Tambahkan nilai MAE ke dalam list
    mae_results.append(mae)
```

```python
my_model_2 = XGBRegressor(n_estimators=100)
my_model_2.fit(X_train, y_train,
               early_stopping_rounds=5,
               eval_set=[(X_valid, y_valid)],
               verbose=False)

predictions_2 = my_model_2.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(predictions_2, y_valid)))
```

```python
my_model_3 = XGBRegressor(n_estimators=100, learning_rate=0.57, n_jobs=4)
my_model_3.fit(X_train, y_train)
```

```python
predictions_3 = my_model_3.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(predictions_3, y_valid)))
```
